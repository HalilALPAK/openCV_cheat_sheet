```python
def apply_threshold(image, thresh_value=127, max_value=255, method=cv.THRESH_BINARY):
    """
    Applies a threshold to the input image.
    :param image: Input image (should be grayscale)
    :param thresh_value: Threshold value
    :param max_value: Maximum value to use with the thresholding
    :param method: Thresholding method (default: cv.THRESH_BINARY)
    :return: Thresholded image
    """
    # If image is not grayscale, convert it
    if len(image.shape) == 3:
        gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
    else:
        gray = image
    ret, thresh_img = cv.threshold(gray, thresh_value, max_value, method)
    return thresh_img

# Example usage:
img = cv.imread('messi.jpg')
thresh_img = apply_threshold(img, thresh_value=100)
cv.imshow('Thresholded Image', thresh_img)
cv.waitKey(0)
cv.destroyAllWindows()
#notebookta gstermek iin bgr->rgb
thresh_img= cv.cvtColor(thresh_img, cv.COLOR_BGR2RGB)
plt.imshow(thresh_img)
plt.axis('off')  # Eksenleri kapat
plt.show()
```

```python
import cv2
import mediapipe as mp
import numpy as np

mp_selfie_segmentation = mp.solutions.selfie_segmentation
cap = cv2.VideoCapture(0)

# Segmentasyon modunu balat (0: selfie iin optimize edilmi)
with mp_selfie_segmentation.SelfieSegmentation(model_selection=1) as selfie_segmentation:
    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Grnty BGR -> RGB evir
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Segmentasyon uygula
        results = selfie_segmentation.process(rgb_frame)

        # Maskeyi al
        mask = results.segmentation_mask
        condition = mask > 0.5  # %50'nin st kii kabul edilir

        # Arka plan blurla
        blurred_frame = cv2.GaussianBlur(frame, (55, 55), 0)

        # Maske ile birletir
        output_image = np.where(condition[..., None], frame, blurred_frame)

        cv2.imshow('MediaPipe Blurred Background', output_image)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np

def live_background_replace(mode=0, background_path=None):
    """
    Kameradan kiiyi alglayp arka plan bulanklatrr (mode=0) veya resimle deitirir (mode=1).
    
    Args:
        mode (int): 0 = blur background, 1 = image background
        background_path (str): Arka plan grselinin yolu (mode=1 iin gerekli)
    """

    mp_selfie_segmentation = mp.solutions.selfie_segmentation
    cap = cv2.VideoCapture(0)

    # Arka plan grselini ykle (mode 1'de)
    if mode == 1 and background_path:
        bg_image = cv2.imread(background_path)
        if bg_image is None:
            print(" Arka plan grseli bulunamad:", background_path)
            return
    else:
        bg_image = None

    with mp_selfie_segmentation.SelfieSegmentation(model_selection=1) as selfie_segmentation:
        while cap.isOpened():
            success, frame = cap.read()
            if not success:
                break

            frame = cv2.flip(frame, 1)  # Aynal grnm
            h, w, _ = frame.shape

            # Gerekirse arka plan resmini yeniden boyutlandr
            if mode == 1 and bg_image is not None:
                resized_bg = cv2.resize(bg_image, (w, h))

            # Segmentasyon maskesi al
            rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = selfie_segmentation.process(rgb_frame)
            mask = results.segmentation_mask
            condition = mask > 0.5  # %50 zeri kii olarak kabul edilir

            if mode == 0:
                # Bulanklatrlm arka plan
                blurred = cv2.GaussianBlur(frame, (55, 55), 0)
                output_image = np.where(condition[..., None], frame, blurred)

            elif mode == 1 and bg_image is not None:
                # Arka plan resimle deitir
                output_image = np.where(condition[..., None], frame, resized_bg)

            else:
                output_image = frame  # Her ihtimale kar orijinal kareyi gster

            cv2.imshow("Kamera - Arka Plan Uygulama", output_image)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

    cap.release()
    cv2.destroyAllWindows()
```

```python
import cv2
import mediapipe as mp
import numpy as np
import pyvirtualcam

def live_virtual_camera(mode=0, background_path=None):
    mp_selfie_segmentation = mp.solutions.selfie_segmentation
    cap = cv2.VideoCapture(0)

    # Arka plan grseli
    if mode == 1 and background_path:
        bg_image = cv2.imread(background_path)
        if bg_image is None:
            print(" Arka plan grseli bulunamad:", background_path)
            return
    else:
        bg_image = None

    with mp_selfie_segmentation.SelfieSegmentation(model_selection=1) as selfie_segmentation:
        success, frame = cap.read()
        if not success:
            print("Kamera alamad!")
            return

        frame = cv2.flip(frame, 1)
        h, w, _ = frame.shape

        # pyvirtualcam balat
        with pyvirtualcam.Camera(width=w, height=h, fps=20) as cam:
            print(f' Sanal kamera alyor: {cam.device}')

            while True:
                success, frame = cap.read()
                if not success:
                    break

                frame = cv2.flip(frame, 1)
                rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                results = selfie_segmentation.process(rgb)
                mask = results.segmentation_mask
                condition = mask > 0.5

                if mode == 0:
                    blurred = cv2.GaussianBlur(frame, (55, 55), 0)
                    output_image = np.where(condition[..., None], frame, blurred)
                elif mode == 1 and bg_image is not None:
                    resized_bg = cv2.resize(bg_image, (w, h))
                    output_image = np.where(condition[..., None], frame, resized_bg)
                else:
                    output_image = frame

                # RGB formatna dntr (virtual camera BGR deil RGB ister)
                cam.send(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
                cam.sleep_until_next_frame()

                # stersen gerek pencere de gsterebiliriz:
                cv2.imshow("nizleme", output_image)
                if cv2.waitKey(1) & 0xFF == ord('q'):
                    break

    cap.release()
    cv2.destroyAllWindows()
```

```python
def custom_hist(gray):
    h,w=gray.shape
    hist=np.zeros([256],dtype=np.int32)
    for row in range(h):
        for col in range(w):
            pv=gray[row,col]
            hist[pv]+=1
    y_pos=np.arange(0,256,1,dtype=np.int32)
    plt.bar(y_pos,hist,align='center',color='r',alpha=0.5)
    plt.xticks(y_pos,y_pos)
    plt.ylabel("sgsg")
    plt.title("histo")

    
def image_hist(image):
    cv.imshow("input",img)
    color=('blue','green','red')
    for i, color in enumerate(color):
        hist=cv.calcHist([img],[i],None,[256],[0,256])
        plt.plot(hist,color=color)
        plt.xlim([0,255])
 
            
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
custom_hist(gray)            
image_hist(img)    
plt.title("beraber")
plt.legend()
plt.show()        
```
